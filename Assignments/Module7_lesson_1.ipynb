```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("company_sales_data.csv")
data.head()
```

```python
plt.plot(data["month_number"], data["facecream"], color="red", ls="--", marker="o")
plt.plot(data["month_number"], data["facewash"], color="purple", ls=":", marker=">")
plt.plot(data["month_number"], data["toothpaste"], color="black", ls="-.", marker="*")
plt.plot(data["month_number"], data["bathingsoap"], color="blue", ls="-", marker="8")
plt.plot(data["month_number"], data["shampoo"], color="violet", marker="H")
plt.plot(data["month_number"], data["moisturizer"], color="yellow")
plt.xlabel("Month of the year")
plt.ylabel("Sales Data")
plt.title("Company Sales Data")
plt.legend(["facecream", "facecream", "toothpaste", "bathingsoap", "shampoo", "moisturizer"])
plt.show()
```
