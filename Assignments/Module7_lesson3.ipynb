```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Compare any two products sales data using a bar chart.

```python
data = pd.read_csv("company_sales_data.csv")
data
```

```python
plt.figure(figsize=(8, 6))
plt.bar(data["month_number"] -0.1, data["facecream"], width= 0.3, color="blue", label="facecream")
plt.bar(data["month_number"] +0.1, data["bathingsoap"], width= 0.3, color="orange", label="bathingsoap")
plt.title("facecream and bathingsoap")
plt.xlabel("Month Number")
plt.ylabel("facecream&bathingsoap")
plt.legend()
plt.savefig("facecream&bathingsoap")
```

Assuming the average monthly sales of bathing soap is 10k, visualize the sales data of bathing soap of all months, showing how they differ from the average sale.


```python
xerr = data["bathingsoap"]/10000
plt.figure(figsize=(8, 6))
plt.bar(range(len(data["month_number"])), data["bathingsoap"], xerr= xerr, color="lightgray", ec = "black", ecolor= "red", capsize=2)
plt.title("sales data of bathing Soap")
plt.xlabel("Month Number")
plt.ylabel("bathingsoap")
plt.savefig("errorbars")
plt.show()
```

Calculate the total sales data for the year of  each product and represent in on a pie chart


```python
data.head()
facecream = data["facecream"].sum()
print(facecream)
facewash = data["facewash"].sum()
toothpaste = data["toothpaste"].sum()
print(toothpaste)
bathingsoap = data["bathingsoap"].sum()
print(bathingsoap)
shampoo = data["shampoo"].sum()
print(shampoo)
moisturizer = data["moisturizer"].sum()
print(moisturizer)

```

```python
data.head()
```

```python
total_sales = [facecream, toothpaste, bathingsoap, shampoo, moisturizer]
product_names = ["facecream", "toothpaste", "bathingsoap", "shampoo", "moisturizer"]

plt.pie(total_sales, labels=product_names, autopct="%d%%")
plt.axis("equal")
plt.title("Total sales data for the year of  each product in percentage")
textprops = {"fontsize": 20}
plt.savefig("piechart")
plt.show()
```
