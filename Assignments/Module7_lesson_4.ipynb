```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data.csv")
data_subset = data[["date","price", "city", "bedrooms", "yr_built", "yr_renovated","sqft_basement"]]

# Distribution of house prices
plt.figure(figsize=(10,6))
plt.hist(x="price", range=(322875.0, 460943.5), data=data_subset)
plt.title("Distribution of house prices".upper(), fontdict={"fontsize":15});
```

```python
# distribution of average house price across the top 10 cities
cities = data_subset.groupby(by="city").price.mean()
cities_reset = cities.reset_index()
top_cities =cities_reset.sort_values("city", ascending=False)
top_cities2 = top_cities.head(10)
top_cities2

plt.figure(figsize=(12,7))
plt.bar(top_cities2["city"], top_cities2["price"])
plt.title("average house price across the top 10 cities".upper(), fontdict={"fontsize":15})

plt.show()
```

```python
# prices of houses from the earliest date to the latest date
# get the needed columns
house_date = data_subset[["date", "price"]]

# convert the date column to a datetime object
house_date["date"] = house_date["date"].apply(pd.to_datetime)

# group the date by price
price_sum = house_date.groupby(by="date", as_index=False)["price"].sum()
plt.figure(figsize=(10,6))
plt.plot(price_sum["date"], price_sum["price"])
plt.xlabel("Date")
plt.ylabel("price")
plt.title("Trend of House Prices", fontdict={"fontsize":15})
plt.show()
```

```python
# The distribution of maximum size bedroom
cities = data_subset.groupby(by="city").bedrooms.max()
cities_reset = cities.reset_index()
cities_reset


top_cities =cities_reset.sort_values("city", ascending=False)
top_cities3 = top_cities.head(10)
top_cities3

plt.figure(figsize=(12,7))
plt.bar(top_cities3["city"], top_cities3["bedrooms"])
plt.title("maximum bedroom size  across the top 10 cities".upper(), fontdict={"fontsize":15})

# plt.show()
# plt.figure(figsize=(8,6))
# plt.bar(cities_reset["city"], cities_reset["bedrooms"])
# plt.show()
```

```python
# distibution of oldest houses

cities = data_subset.groupby(by="city").yr_built.max()
cities_reset = cities.reset_index()
cities_reset

top_cities =cities_reset.sort_values("city", ascending=False)
top_cities4 = top_cities.head(10)
top_cities4

plt.figure(figsize=(12,7))
plt.bar(top_cities4["city"], top_cities4["yr_built"])
plt.title("oldest houses across top 10 cities".upper(), fontdict={"fontsize":15})

# plt.show()
# # unique_cities= data["city"].unique()
# plt.bar(cities_reset["city"], cities_reset["yr_built"])
```

```python
# last renovated
# cities = data_subset.groupby(by="city").yr_renovated.max()
# cities_reset = cities.reset_index()
# cities_reset
# plt.bar(cities_reset["city"], cities_reset["yr_renovated"])
sns.kdeplot(x=data_subset["yr_renovated"], data = data_subset, shade=True)
plt.title("Distribution of years Houses were Last Renovated".upper(), fontdict={"fontsize":15});
```

```python
# Distribution of Average basement square feets across top 10 cities

cities = data_subset.groupby(by="city").sqft_basement.mean()
cities_reset = cities.reset_index()
cities_reset

top_cities =cities_reset.sort_values("city", ascending=False)
top_cities5 = top_cities.head(10)
top_cities5

plt.figure(figsize=(12,7))
plt.bar(top_cities5["city"], top_cities5["sqft_basement"])
plt.title("Average basement square feets across top 10 cities".upper(), fontdict={"fontsize":15});

```

```python
# relationship between prices and numbers of bedrooms
plt.figure(figsize=(10,6))
sns.histplot(x=data_subset["bedrooms"], y=data_subset["price"], data=data_subset, color="red");
plt.title("prices and numbers of bedrooms".upper(), fontdict={"fontsize":15})
```

```python
# The relationship bwtween number of bedrooms and year the house was built
sns.histplot(x=data_subset["bedrooms"], y=data_subset["yr_built"], data=data_subset);
plt.title("numbers of bedrooms and year house was built".upper(), fontdict={"fontsize":15});
```
