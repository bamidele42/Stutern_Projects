```python
orders["order_id"] = orders["order_id"].astype(object)
orders["order_id"].dtypes
```

```python
# inner merge operation between the prders table and products table
inner_merge_1 = pd.merge(orders, product_dataset, how="inner", on= "product_id" )
inner_merge_1.set_index("order_id")
```

```python
# inner merge of customers table with the result of the above operation
inner_merge_2 = inner_merge_1.merge(customer_dataset)
inner_merge_2.set_index("order_id")
```

## Observations
The first thing that was done: was to change the type of the "order_id" column to an object type to avoid a ValueError"You are trying to merge on object and int64 columns. If you wish to proceed you should use pd.concat"

Then the orders table and the product product table were merged base on the "product_id" (common to both dataframes) column using an inner merge, it also default to inner merge if none was specified. Notice that the columns from the product_dataset has been added to the inner_merge_1 table. The inner_merge_2 contain columns from the customer dataset set after which a inner join has been performed based on the result of the previous output.

```python
# inner merge between customers table and orders table
inner_merge_3 = pd.merge(customer_dataset, orders, left_on="customer_id", right_on="customer_id")
inner_merge_3
```

## Observation:
The above table shows the result for the inner merge done base on the customer dataset and the order's dataset, the result contains columns from the customer and order dataset base on the right_on keyword and the left_on keyword that were passed in as arguments. 

```python
# Outer merge between customers and product tables
outer_merge = customer_dataset.merge(product_dataset, how="outer")
outer_merge
```

## Observation:
The customer and product dataset do not have a common key on which merge operation can be performed, (MergeError: No common columns to perform merge on. Merge options: left_on=None, right_on=None, left_index=False, right_index=False). One way to resolve this issues is to create a customer_id column in the product table. The both tables were merged based on the newly added column to the product table. 

```python
product_dataset["customer_id"] = ["2001", "2001", "2002", "2003", "2004", "2004", "2005", "2005", "2006", "2007", "2008", "2008", "2009", "2009", "2010"]
#product_dataset
outer_merge = customer_dataset.merge(product_dataset, how="outer")
outer_merge
```

```python
# left merge between orders and customer tables
left_merge = orders.merge(customer_dataset, how="left")
left_merge
```

## Observation
The result above shows that the merge happpened based on the customer_id found in the left dataframe (orders table). the column is also present in the customers table

```python
# right merge between orders and customers tables
right_merge = product_dataset.merge(customer_dataset, how="right")
right_merge
```

## Observation
The result above shows that the merge happpened based on the customer_id found in the right dataframe (product table). the column is also present in the customers table

```python
concatenation = pd.concat([orders, product_dataset, customer_dataset])
concatenation.dropna(axis=1, how="any")
```
