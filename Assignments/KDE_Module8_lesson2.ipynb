```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset("tips")
data.head()
```

```python
sns.kdeplot(data["size"], shade=True)
plt.title("size Feature of the tip dataset".upper());
```

The chart above indicated that the size of the tips are clustered around 2, tip size around 3 and 4 also has some frequent values while tips the that has the values of 1, 5, 6 and 7 has the least tip size

```python
sns.boxplot(data=data, x="sex", y="total_bill");
```

The chart above shows the male higher total bill than female, 75th quartile of the male total bill falls within 25, while that of female falls within 21, 25th quartile of the total bill for male within 15 while that of female falls within 14 the is no much variation in the lower percentiles, the male gender has higher outliers compare to the female gender.

```python
sns.boxplot(data=data,x="day", y="tip");
```

The chart above above shows that sunday has tips generated frequency, with the highest upper quartile of about 4, while thursday, friday and saturday has the same upper quartile values, the middle quartile or median values varies. Thurday and saturday has values that are far from the distribution (outlier)

```python
sns.boxplot(data=data, x="day", y="total_bill", hue="smoker");
```

The chart above shows that there smokers give more tips on sundays and saturdays, non smokers gives more tips on fridays acorrding to the chart, with about 23 on the upper quartile range. smokers and non smokers, have almost thesame upper quartile range on thursdays but there is a slight variation in the lower quartile range. saturdays has the highest numbers of outliers in the dataset.

```python
sns.violinplot(data=data, x="total_bill", y="time", hue="sex");
```

The violin plot above shows a slight variation in the tips that was given during dinner by both male and female with about 25 & 24 total bills respectively, there is more variation in the totall bills for lunch tips with upper quartile ranging from 20 to 25 for female and male respectively

```python
sns.violinplot(data=data, x="tip", y="day", hue="time");
```

The chart above shows that more tips were given on thurday's lunch time, the tips are usually more evenly spread around the dinner time, the dinner time has higher middle values than the lunch time, there were no tip given on saturdays and sundays during lunch time, but there were tips on satudays and sundays during dinner time
