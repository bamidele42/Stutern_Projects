The missing values in the "RATING" column was filled with avarage of all the values in the column

```python
data["RATING"].fillna(data["RATING"].mean(), inplace = True)
data.head()
```

The missing values in the "RunTime" column was filled with average of all the values in the column, and the values were replaced permanently in original dataset using the inplace=True argument.

```python
data["RunTime"].fillna(data["RunTime"].mean(), inplace = True)
data.tail(20)
```

The Gross missing values in the groos column were replaced with Zeros, notice that the 0 is in string format, this to align with the object data type of the Gross column

```python
data["Gross"].fillna("0", inplace=True)
data.reset_index()
data.head(20)
```
