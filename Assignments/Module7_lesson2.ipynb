```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("anscombe_quartet_data.csv")
data
```

## First subplots

```python
plt.figure(figsize=(8,6))
plt.subplot(2, 2, 1)
plt.title("subplot x1y1")
plt.plot([4, 14], [4, 12])
plt.scatter(data["x1"], data["y1"], color="red")
plt.xlabel("x1")
plt.ylabel("y1")

plt.subplot(2, 2, 2)
plt.title("subplot x2y2")
plt.plot([4, 14], [4, 12])
plt.scatter(data["x2"], data["y2"], color="black")
plt.xlabel("x2")
plt.ylabel("y2")

plt.subplot(2, 2, 3)
plt.title("subplot x3y3")
plt.plot([4, 14], [4, 13])
plt.scatter(data["x3"], data["y3"], color="blue")
plt.xlabel("x3")
plt.ylabel("y3")

plt.subplot(2, 2, 4)
plt.title("subplot x4y4")
plt.plot([4, 20], [4, 12])
plt.scatter(data["x4"], data["y4"], color="green")
plt.xlabel("x4")
plt.ylabel("y4")

plt.suptitle("PLOT 1")
plt.subplots_adjust(wspace=0.3, hspace=0.3, bottom=1, top=2)
```

## Second subplots

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(8,6))

ax1.plot([4, 14], [4, 13])
ax1.scatter(data["x1"], data["y1"], c= "red")
ax1.set_title("x1y1")



ax2.plot([4, 14], [4, 13])
ax2.scatter(data["x2"], data["y2"], c= "black")
ax2.set_title("x2y2")

ax3.plot([4, 14], [4, 13])
ax3.scatter(data["x3"], data["y3"], c= "blue")
ax3.set_title("x3y3")

ax4.plot([4, 20], [4, 12])
ax4.scatter(data["x4"], data["y4"], c= "green")
ax4.set_title("x4y4")

plt.suptitle("PLOT 2")
plt.subplots_adjust(wspace=0.3, hspace=0.3, bottom=1, top=2)
```

## Third Subplots

```python
fig = plt.figure(figsize=(8, 6))

ax5 = fig.add_subplot(221, title=("x1y1"))
ax5.plot([4, 14], [4, 13])
ax5.scatter(data["x1"], data["y1"], c="red")

ax6 = fig.add_subplot(222, title=("x2y2"))
ax6.plot([4, 14], [4, 13])
ax6.scatter(data["x2"], data["y2"], c="black")

ax7 = fig.add_subplot(223, title=("x3y3"))
ax7.plot([4, 14], [4, 13])
ax7.scatter(data["x3"], data["y3"], c="blue")

ax8 = fig.add_subplot(224, title=("x1y1"))
ax8.plot([4, 20], [4, 12])
ax8.scatter(data["x4"], data["y4"], c="green")

plt.suptitle("PLOT 3")
plt.subplots_adjust(wspace=0.3, hspace=0.3, bottom=1, top=2)
```
